```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
temperature=np.random.randint(20,40,50)
sales=2.5*temperature+np.random.randint(0,5,50)
df=pd.DataFrame({'Temperature':temperature,'Sales':sales})
print(df.head())
```

```python
#model
X=df[['Temperature']]
Y=df['Sales']
model=LinearRegression()
model.fit(X,Y)
```

```python
#plot
plt.scatter(X,Y)
plt.plot(X,model.predict(X),color='skyblue')
plt.xlabel('Temperature(°c)')
plt.ylabel('Sales')
plt.title('Simple Linear regression')
plt.show()
```
