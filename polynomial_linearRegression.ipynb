```python
#Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
#Load Dataset
url="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df=pd.read_csv(url)
df.head()
```

```python
#prepare data
x=df[["Hours"]].values
y=df[["Scores"]].values
```

```python
#Polynomial Features Transfromation
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
```

```python
#train the model
model=LinearRegression()
model.fit(x_poly,y)
```

```python
y_pred=model.predict(x_poly)

print("MSE:",mean_squared_error(y,y_pred))
print("R2 Score:",r2_score(y,y_pred))
```

```python
# Step 7: Plotting the results
plt. scatter(x, y, color='blue', label='Actual data')
# Sorting for smooth curve
x_sorted = np.sort(x, axis=0)
y_sorted_pred = model. predict (poly. transform (x_sorted) )
plt. plot(x_sorted, y_sorted_pred, color='red', label='Polynomial Regression' )
plt. title ('Polynomial Regression (Hours vs Scores) ' )
plt. xlabel ( 'Hours Studied')
plt. ylabel ('Score' )
plt. legend ()
plt. show ()
```
