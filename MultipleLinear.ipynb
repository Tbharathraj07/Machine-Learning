```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(1)
Study_hours=np.random.randint(1,10,50)
attendance=np.random.randint(50,100,50)
sleep_hours=np.random.randint(5,9,50)
score=(5*Study_hours+0.5*attendance+2*sleep_hours)+np.random.normal(0,10,50)
```

```python
df=pd.DataFrame({
    'Study_hours':Study_hours,
    'attendance':attendance,
    'slepp_hours':sleep_hours,
    'score':score
})
print(df)
```

```python
#model
X=df[['Study_hours','attendance','slepp_hours']]
Y=df['score']
model=LinearRegression()
model.fit(X,Y)
```

```python
print("Coefficients:",model.coef_)
print("Intercept:",model.intercept_)
```
